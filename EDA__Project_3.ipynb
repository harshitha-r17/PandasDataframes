```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df1=pd.read_csv("/content/amsterdam_weekdays.csv")
df2=pd.read_csv("/content/amsterdam_weekends.csv")
df3=pd.read_csv("/content/athens_weekdays.csv")
df4=pd.read_csv("/content/athens_weekends.csv")
df5=pd.read_csv("/content/berlin_weekdays.csv")
df6=pd.read_csv("/content/berlin_weekends.csv")
df7=pd.read_csv("/content/budapest_weekdays.csv")
df8=pd.read_csv("/content/budapest_weekends.csv")
df9=pd.read_csv("/content/barcelona_weekdays.csv")
df10=pd.read_csv("/content/barcelona_weekends.csv")
df11=pd.read_csv("/content/lisbon_weekdays.csv")
df12=pd.read_csv("/content/lisbon_weekends.csv")
df13=pd.read_csv("/content/london_weekdays.csv")
df14=pd.read_csv("/content/london_weekends.csv")
df15=pd.read_csv("/content/paris_weekdays.csv")
df16=pd.read_csv("/content/paris_weekends.csv")
df17=pd.read_csv("/content/rome_weekdays.csv")
df18=pd.read_csv("/content/rome_weekends.csv")
df19=pd.read_csv("/content/vienna_weekdays.csv")
df20=pd.read_csv("/content/vienna_weekends.csv")
```

```python
df1["city"]="Amsterdam"
df3["city"]="Athens"
df5["city"]="Berlin"
df7["city"]="Budapest"
df9["city"]="Barcelona"
df11["city"]="Lisbon"
df13["city"]="London"
df15["city"]="Paris"
df17["city"]="Rome"
df19["city"]="Vienna"
df2["city"]="Amsterdam"
df4["city"]="Athens"
df6["city"]="Berlin"
df8["city"]="Budapest"
df10["city"]="Barcelona"
df12["city"]="Lisbon"
df14["city"]="London"
df16["city"]="Paris"
df18["city"]="Rome"
df20["city"]="Vienna"
```

```python
# Concatenate files into two master datasets: one for weekdays and one for weekends.
weekdays=pd.concat([df1,df3,df5,df7,df9,df11,df13,df15,df17,df19],ignore_index=True)
weekdays.head(5)
```

```python
weekends=pd.concat([df2,df4,df6,df8,df10,df12,df14,df16,df18,df20],ignore_index=True)
weekends.head(5)
```

```python
# Combine the datasets into a single unified dataset with an additional column indicating "day_type" as either 'weekday' or 'weekend'.
weekends["data_type"]="weekend"
weekdays["data_type"]="weekdays"
```

```python
combineddata=pd.concat([weekdays,weekends],ignore_index=True)
combineddata.head(5)
```

```python
#Data Cleaning:
#Check for missing values and impute or remove them appropriately.
combineddata.isnull().sum()
```

```python
# Handle duplicates if present.
combineddata.duplicated().sum()
```

```python
#Address any outliers in key features such as realSum (price) and guest_satisfaction_overall.
fig, axes = plt.subplots(3, 2,figsize=(10,10))

sns.boxplot(data=combineddata, x='realSum', ax=axes[0,0])
sns.boxplot(data=combineddata, x='cleanliness_rating', ax=axes[0,1])
sns.boxplot(data=combineddata, x='guest_satisfaction_overall', ax=axes[1,0])
sns.boxplot(data=combineddata, x='dist', ax=axes[1,1])
sns.boxplot(data=combineddata, x='metro_dist', ax=axes[2,1])
plt.show()
```

```python
#Exploratory Analysis-Univariate Analysis
#Price Distribution: Plot histograms and boxplots of realSum to understand price distributions for weekdays and weekends separately.
sns.barplot(data=combineddata, x='data_type', y='realSum')
```

```python
# Room Types: Visualize counts of different room_type categories (e.g., shared, private, entire home).
fig, axes = plt.subplots(3, 2,figsize=(10,10))

sns.countplot(data=combineddata, x='room_type', ax=axes[0,0])
sns.countplot(data=combineddata, x='room_shared', ax=axes[0,1])
sns.countplot(data=combineddata, x='room_private', ax=axes[1,0])
sns.countplot(data=combineddata, x='person_capacity', ax=axes[1,1])
sns.countplot(data=combineddata, x='cleanliness_rating', ax=axes[2,0])
sns.countplot(data=combineddata, x='bedrooms', ax=axes[2,1])
plt.show()
```

```python
# Guest Satisfaction: Plot distribution of guest_satisfaction_overall ratings.
sns.histplot(data=combineddata, x='guest_satisfaction_overall',kde=True)
```

```python
#Price vs. Room Type: Compare average prices for different room_type categories between weekdays and weekends using bar plots.
sns.barplot(data=combineddata, x='room_type', y='realSum', hue='data_type')
```

```python
#Superhost Status: Analyze how host_is_superhost impacts average prices and guest satisfaction ratings using boxplots or violin plots.
sns.barplot(data=combineddata, x='host_is_superhost', y='realSum')
```

```python
sns.barplot(data=combineddata, x='host_is_superhost', y='guest_satisfaction_overall')
```

```python
# Distance Impact: Plot realSum against dist and metro_dist to understand how distance from city center and metro stations affects pricing.
sns.scatterplot(data=combineddata, x='dist', y='realSum')
```

```python
sns.lineplot(data=combineddata, x='metro_dist', y='realSum')
```

```python
#Multivariate Analysis
correlation=combineddata[['realSum','cleanliness_rating','guest_satisfaction_overall','person_capacity','bedrooms','dist','metro_dist','attr_index','attr_index_norm','rest_index','rest_index_norm','lng','lat']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm',fmt="0.2f")
```

```python
#Feature Interactions: Explore how combinations of features (e.g., multi and biz) influence price and satisfaction.
sns.barplot(data=combineddata, x='multi', y='realSum', hue='biz')
```

```python
sns.barplot(data=combineddata, x='multi', y='guest_satisfaction_overall', hue='biz')
```
